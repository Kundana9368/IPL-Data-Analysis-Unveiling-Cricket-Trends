```python
teams = match['toss_winner'].unique()
decision_making = pd.DataFrame(columns=['Toss Winner', 'Decision', 'Times'])

for team in teams:
    # Counting the number of times the team won the toss and chose to bat
    bat_count = match[(match['toss_winner'] == team) & (match['toss_decision'] == 'bat')].shape[0]
    
    # Counting the number of times the team won the toss and chose to field
    field_count = match[(match['toss_winner'] == team) & (match['toss_decision'] == 'field')].shape[0]
    
    # Create temporary DataFrames for bat and field decisions
    bat_df = pd.DataFrame({'Toss Winner': [team], 'Decision': ['bat'], 'Times': [bat_count]})
    field_df = pd.DataFrame({'Toss Winner': [team], 'Decision': ['field'], 'Times': [field_count]})
    
    # Concatenate the temporary DataFrames to the main DataFrame
    decision_making = pd.concat([decision_making, bat_df, field_df], ignore_index=True)

print(decision_making)

```

```python
# Plotting
sns.catplot(x='Toss Winner', y='Times', hue='Decision', kind='bar', data=decision_making, height=8, aspect=2)
plt.xticks(rotation=90)
plt.title('Decisions of Teams Who Won The Toss', fontsize=16)

ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python
# Group by batsman and aggregate the sum of runs...top15batsmen
top_batsman = delivery.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values('batsman_runs', ascending=False).head(15)
top_batsman.set_index('batsman', inplace=True)
ax = top_batsman.plot(kind='barh')

# Add data labels to the bars
for i, v in enumerate(top_batsman['batsman_runs']):
    ax.text(v + 5, i, str(v), color='black')  # Adjust the position of the text as needed

plt.title('Batsman Who Scored The Most Runs', fontsize=16)
plt.xlabel("Total Runs")
plt.ylabel("Batsman")
plt.show()
```

```python
#most runs by a bowler
total_runs=delivery.groupby('bowler')['total_runs'].agg('sum').reset_index().sort_values('total_runs', ascending=False).head(15)
total_runs.set_index('bowler',inplace=True)
ax=total_runs.plot(kind='barh')

# Add data labels to the bars
for i, v in enumerate(total_runs['total_runs']):
    ax.text(v + 5, i, str(v), color='black')  # Adjust the position of the text as needed

plt.title('Bowlers Who Conceded The Most Runs to Other Teams', fontsize=16)
plt.xlabel("Total Runs")
plt.ylabel("Batsman")
plt.show()
```

```python
#PP Chawla gave the most runs to other teams batsmen. Runs he gave to each team

runs_conceded = delivery['bowler'] == 'PP Chawla'
runs_given = delivery[runs_conceded].groupby('batting_team')['total_runs'].sum()
runs_given.plot(kind='bar')

plt.title('Runs Conceded by PP Chawla to Each Team', fontsize=16)
plt.xlabel('Opposing Team')
plt.ylabel('Total Runs Conceded')
plt.xticks(rotation=90)
plt.show()
```

```python
#Since Virat Kohli is the best batsman with the most runs
kohli_runs = delivery[delivery['batsman'] == 'V Kohli']

# Group by bowler and aggregate the sum of runs scored by V Kohli against each bowler
runs_scored_by_bowler = kohli_runs.groupby('bowler')['batsman_runs'].sum()
top_10_bowlers = runs_scored_by_bowler.nlargest(10)

top_10_bowlers.plot(kind='bar')
plt.title('Top 10 Bowlers Against Whom V Kohli Has Scored the Most Runs', fontsize=16)
plt.xlabel('Bowler')
plt.ylabel('Total Runs Scored by V Kohli')
plt.xticks(rotation=45)
plt.show()
```

```python
#best batsman and bowler based on the runs and wickets taken respectively

# Determine the best batsman for each team based on highest runs
best_batsman = delivery.groupby('batting_team').apply(lambda x: x.groupby('batsman')['batsman_runs'].sum().idxmax())
# Determine the best bowler for each team based on most wickets
best_bowler = delivery.groupby('bowling_team').apply(lambda x: x[x['dismissal_kind'] != 'run out'].groupby('bowler')['dismissal_kind'].count().idxmax())


team_summary = pd.DataFrame({'Best Batsman': best_batsman, 'Best Bowler': best_bowler})
print(team_summary)
```

```python
#average runs per over

delivery_by_over=delivery[runs_conceded]
delivery_by_over=delivery_by_over[['batting_team','over','batsman_runs']]
x=delivery_by_over.pivot_table(values='batsman_runs',index='batting_team',columns='over',aggfunc='count')
sns.heatmap(x,cmap='summer')
```

```python
#Top dismissal kind

top_dismissal_kind = delivery['dismissal_kind'].value_counts().idxmax()
# Plot the count of dismissal kinds
sns.countplot(x='dismissal_kind', data=delivery)
plt.xticks(rotation='vertical')
plt.title('Top Dismissal Kind: ' + top_dismissal_kind)
plt.show()
```

```python
#Top 10 Bowlers with Dismissals via Catch, Bowled, LBW, Caught & Bowled, and Stumped

mask1=delivery['dismissal_kind']=='caught'
mask2=delivery['dismissal_kind']=='bowled'
mask3=delivery['dismissal_kind']=='lbw'
mask4=delivery['dismissal_kind']=='caught and bowled'
mask5=delivery['dismissal_kind']=='stumped'

new_del=delivery[mask1 | mask2 | mask3 | mask4 | mask5]
new_del.groupby('bowler')['player_dismissed'].agg('count').sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
#number of runs scored by a batsman against a specific bowler
mask=delivery['bowler']=='JJ Bumrah'
mask2=delivery['batsman']=='V Kohli'
delivery[mask].groupby('batsman')['batsman_runs'].agg('count').sort_values(ascending=False)['V Kohli']
```

```python
#Batsmen who scored the most number of sixes

temp_df = delivery.groupby('batsman')['batsman_runs'].agg(lambda x: (x == 6).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).head(10).reset_index(drop=True)
plt.figure()
ax = sns.barplot(x='batsman', y='batsman_runs', data=temp_df)

# Adding data labels to the bars
for i, v in enumerate(temp_df['batsman_runs']):
    ax.text(i, v + 0.2, str(v), ha='center', va='bottom', fontsize=10)
plt.title("Most Sixes", fontsize=16)
plt.xlabel("Batsman", fontsize=14)
plt.ylabel("Number of 6's", fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

```python
temp_df = delivery.groupby('bowler')['is_wicket'].agg('sum').reset_index().sort_values(by='is_wicket', ascending=False).reset_index(drop=True).head(10)
plt.figure()
ax = sns.barplot(x='bowler', y='is_wicket', data=temp_df)

# Adding data labels to the bars
for i, v in enumerate(temp_df['is_wicket']):
    ax.text(i, v + 0.2, str(v), ha='center', va='bottom', fontsize=10)
plt.title("Most Wickets by a Bowler", fontsize=16)
plt.xlabel("Bowler", fontsize=14)
plt.ylabel("Number of Wickets", fontsize=14)
plt.xticks(rotation=90)
plt.show()
```
